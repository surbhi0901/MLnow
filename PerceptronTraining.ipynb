#Problem Statement

Train a  model to learn the table of 10 

Equation:y=10*x

#Data creation

Equation: y=10x

x=[0,1,2,3,4,5,...]

y=[0,10,20,30,40,50....]

```python
x=[i for i in range(21)]
print (x)
y=[i*10 for i in range(21)]
print(y)
```

#Approach-1

Defining a function for the equation y=10x


```python
def tableof10(x):
  y= 10*x
  return y

for i in x:
  print(tableof10(i),end=' ')
```

#Approach-2

I am super enthusiastic about so I am gonna make a ML model for table of 10

Now, we will need xTrain, yTrain, xTest and yTest from x and y.

```python
print("This is x: ",x)
print(f'This is y:{y}') ##this is fstring!!!
```

```python
xTrain=x[:-5] # TRAINING DATA
yTrain=y[:-5] #TRAINING LABELS

xTest=x[-5:] #TESTING DATA
yTest=y[-5:] #TESTING LABEL

# WE WILL PRINT USING FSTRING:

print(f'''
Training data:
xTrain: {xTrain}
yTrain: {yTrain}

Testing data:
xTest: {xTest}
yTest: {yTest}
''')
```

###Now we will start coding using TENSORFLOW:

```python
#Import required libraries

import tensorflow as tf
from tensorflow import keras
```

### Defining the model:

```python
model=tf.keras.Sequential([tf.keras.layers.Dense(units=1,input_shape=[1])])
```

###Compiling the model to reduce error

```python
model.compile(optimizer='adam',loss='mae') #mae is mean absolute error
```

```python
nEpoch=5000 #THIS IS THE NUMBER OF TIMES FOR WHICH WE TRAIN OUR DATASET
model.fit(x=xTrain, y=yTrain, validation_data=(xTest,yTest), epochs=nEpoch)
```
